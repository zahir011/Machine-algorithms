# Naive Bayes:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lod the dataset:
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
# Input and Output:
x = df.iloc[:, :-1]   # Features
y = df.iloc[:, -1:]   # Labels
```

```python
x.head()
```

```python
y.head()
```

```python
# Training and fitting the model:
from sklearn.naive_bayes import GaussianNB
model = GaussianNB().fit(x,y)
model
```

```python
# Train,test, and split and checking the accuracy
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=451)
```

```python
# Training the model on training data:
from sklearn.naive_bayes import GaussianNB
model = GaussianNB().fit(x_train,y_train)

# making the prediction on the test set:
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
score = metrics.accuracy_score(y_test,y_pred)
print('Gaussian Naive Bayes model Accuaracy(in %): ', metrics.accuracy_score(y_test,y_pred)*100)
```

```python
# Confussion metrix:
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (12,12))
sns.heatmap(cm, annot=True, fmt='.3f', linewidth=.5, square = True, cmap='Spectral')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Gaussian Naive Bayes Model accuracy(in %): {0}'.format(score*100)
plt.title(all_sample_title, size=15)
```
