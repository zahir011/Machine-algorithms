```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt

# import dataset:

dataset = pd.read_csv('Position_Salaries.csv')
dataset.head()
```

```python
# Extract features and target variable
x = dataset["Level"].values.reshape(-1, 1)  # Reshape to 2D array for sklearn
y = dataset["Salary"].values  # Target variable
```

```python
x = dataset.iloc[: 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Splitting the data into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Fit the linear regression to the dataset:
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Extract features and target variable
x = dataset["Level"].values.reshape(-1, 1)  # Reshape to 2D array for sklearn
y = dataset["Salary"].values  # Target variable

# Check the shapes of x and y
x.shape, y.shape


lin_reg = LinearRegression()
lin_reg.fit(x, y)

# Visualize the linear regression result:
def viz_linear():
    plt.scatter(x, y, color='red')
    plt.plot(x, lin_reg.predict(x), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()

# Call the visualization function
viz_linear()
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Create polynomial features
poly_features = PolynomialFeatures(degree=4)
x_poly = poly_features.fit_transform(x)

# Fit Polynomial Regression model
poly_model = LinearRegression()
poly_model.fit(x_poly, y)

# Visualization function
def viz_polynomial():
    plt.scatter(x, y, color='red')  # Scatter plot of actual data
    plt.plot(x, poly_model.predict(x_poly), color='blue')  # Polynomial regression curve
    plt.title('Truth or Bluff (Polynomial Regression)')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.show()

# Call the visualization function
viz_polynomial()
```

```python
# predict a new result with linear regression:
pred_linear = lin_reg.predict([[11]])
pred_linear
```

```python
# Predicting a new result with Polynomial Regression
new_input = np.array([[11]])  # Position level 11
new_input_poly = poly_features.transform(new_input)  # Transform using PolynomialFeatures
pred_polynomial = poly_model.predict(new_input_poly)  # Predict using the trained model

# Output the predicted salary
pred_polynomial
```
