```python
# Step-1: Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# import the dataset:
df = pd.read_csv('Salary Data22.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
x = df[['Age','YearsExperience']]
y = df['Salary']
```

```python
model = LinearRegression().fit(x,y)
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5,30]])
```
